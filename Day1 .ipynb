```python
df3 = pd.DataFrame({"x":[1, 2, 3, 4, 5], "y":[10, 20, 30, 20, 10], "z":[5, 4, 3, 2, 1]})
df3.iplot(kind='surface', colorscale='ylorbr') #, x='x', y='y', z='z'
```
